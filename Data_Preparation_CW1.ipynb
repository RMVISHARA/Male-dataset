```python
# Find the index of the row with age 190
index_to_drop = df[df['AGE_years'] == 190].index

# Drop the row from the dataframe
df.drop(index_to_drop, inplace=True)
```

```python
# Multiply records with values less than 2 by 100
df.loc[df['HEIGHT_cm'] < 2, 'HEIGHT_cm'] *= 100
```

```python
from sklearn.impute import SimpleImputer

# Calculate mean blood pressure for each age category
age_groups = df.groupby('AGE_years')
mean_bp_by_age = age_groups['SYSTOLIC_BLOOD_PRESSURE_mmHg'].mean()

# Create a function to impute missing values with mean blood pressure by age
def impute_bp(row):
    if pd.isnull(row['SYSTOLIC_BLOOD_PRESSURE_mmHg']):
        return mean_bp_by_age[row['AGE_years']]
    else:
        return row['SYSTOLIC_BLOOD_PRESSURE_mmHg']

# Apply the imputation function to the 'SYSTOLIC_BLOOD_PRESSURE_mmHg' column
imputer = SimpleImputer(strategy='mean')
df['SYSTOLIC_BLOOD_PRESSURE_mmHg'] = df.apply(impute_bp, axis=1)
print(df.apply(impute_bp, axis=1))

```

```python

# Replace blank values with 0 in the 'CIGARETTES_CONSUMED_PER_DAY' column
df['CIGARETTES_CONSUMED_PER_DAY'].fillna(0, inplace=True)
```

```python

# Calculate the mean computer use time per day for each age group
mean_computer_use_time = df.groupby('AGE_years')['COMPUTER_USE_TIME_PER_DAY_HOURS'].mean()

# Replace values greater than 24 with the mean computer use time for that age group
for age, mean_time in mean_computer_use_time.items():
    df.loc[(df['AGE_years'] == age) & (df['COMPUTER_USE_TIME_PER_DAY_HOURS'] > 24), 'COMPUTER_USE_TIME_PER_DAY_HOURS'] = np.nan
    imp = SimpleImputer(missing_values=np.nan, strategy='mean')
    df.loc[df['AGE_years'] == age, 'COMPUTER_USE_TIME_PER_DAY_HOURS'] = imp.fit_transform(df.loc[df['AGE_years'] == age, ['COMPUTER_USE_TIME_PER_DAY_HOURS']])
```

```python

# Make negative values in the 'Visceral_Fat_Volume_Litres' column positive
df['Visceral_Fat_Volume_Litres'] = df['Visceral_Fat_Volume_Litres'].abs()
print(df['Visceral_Fat_Volume_Litres'].abs())
```

```python
df['SMOKING_STATUS'] = df['CIGARETTES_CONSUMED_PER_DAY'].apply(lambda x: 1 if x > 0 else 0)
print(df['CIGARETTES_CONSUMED_PER_DAY'].apply(lambda x: 1 if x > 0 else 0))
```
